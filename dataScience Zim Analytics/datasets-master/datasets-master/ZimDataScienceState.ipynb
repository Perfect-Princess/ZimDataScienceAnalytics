```python
chart2 = data['AI/ML/Ds'].value_counts().rename_axis('unique_values').reset_index(name = 'counts')
```

```python
chart2
```

```python
chart2.info()
```

```python
df_numeric = chart2[['counts']]
```

```python
df_numeric
```

```python
df_categorical = chart2[['unique_values']]
```

```python
df_categorical
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder =LabelEncoder()
encoder.fit(df_categorical['unique_values'])
```

```python
unique_values = encoder.transform(df_categorical['unique_values'])
```

```python
unique_values = pd.DataFrame(unique_values,columns = ['unique_values'])
```

```python
new_df = pd.concat([df_numeric,unique_values],axis =1)
```

```python
new_df
```
